## Introduction

The analysis below is that of a data set called "No show", this dataset collects information from 100k medical appointments in Brazil and is focused on the question of whether or not patients show up for their appointment. A number of characteristics about the patients were collected and they are listed below;
1.AppointmentDay:This is the actual day of patient's appointment
2.Age:Patients age
3.Neibourbourhood:The neighborhood where the appointment takes place
4.Scholarship: whether or not the patient is enrolled in the brazillian Welfare Program Bolsa Familia[1- True, 0- False]
5.Hypertension: wherther or not the patient is hypertensive
6.Diabetes:Whether or not the patient is diabetic[1-True, 0- False]
7.Alcoholism: Whether or not the patient is an alcoholic [1-TRUE, 0-False]
8.Handcap:Whether or not the patient is an handicap[1-True,0-False]
9.SMS_received:Whether or not the patient received a SMS reminding them of their appointment [1-True, 0-False]
10.NO-show: Whether or not the patient showed up eventually for the appointment[No-"showed p", Yes-"Did not show up"]

## The idea behind this project is to be able to determine the factors or reasons patients refused to show up for their appointment.

### Questions for Analysis
1.Does patient's gender affect showing up?
2.Does been handicapped affect showing up?
3.Does the Scholarship enrollment affect showing up?
4.Does been hypertensive affect showing up?
5.Does being alcoholic affect showing up?

```python
def plot_bar(x,title,xlabel,ylabel):
    """A function to draw a bar chat"""
    x.plot.bar();
    plt.xlabel(xlable)
    plt.title(title)
    plt.ylabel(ylabel)
    
```

```python
#This code plots the graph of the percentage of each Gender that showed up and don't that didn't show up for thier appointment

data = round(pd.DataFrame(df.groupby('Gender')['No_show'].value_counts(normalize=True)) *100)
plot_bar( data,
             "Percentage of availability and Non-availablity of each gender",
            "Gender and their availability status",
             "percentage of availability" )
```

## The barchart above is used to determine the percentage of availability of both genders in my data set.
